```python
def date_creator():
  for i in range(1,51):
    lnk = "https://www.hurriyetdailynews.com/search/science?p={}".format(str(i))
    science_page_link.append(lnk)
  for i in range(1,51):
    lnk = "https://www.hurriyetdailynews.com/search/technology?p={}".format(str(i))
    technology_page_link.append(lnk)
```

```python
def technology_get_link_creator(url):
  html = requests.get(url).content                                                          
  soup = BeautifulSoup(html,"html.parser")
  link_list = soup.find_all("div",{"class" : "col-md-4 col-sm-4"})
  for i in link_list:
    href = i.find("a").get("href")
    if href.startswith("https://www.hurriyetdailynews.com"):
      pass
    else:
      href = "https://www.hurriyetdailynews.com"+href
      technology_links
      with open("daily_hurriyet_technology_Link.txt","a",encoding="utf-8") as file :
        file.write(href+"\n")

```

```python
def science_get_link_creator(url):
  html = requests.get(url).content                                                          
  soup = BeautifulSoup(html,"html.parser")
  link_list = soup.find_all("div",{"class" : "col-md-4 col-sm-4"})
  for i in link_list:
    href = i.find("a").get("href")
    if href.startswith("https://www.hurriyetdailynews.com"):
      pass
    else:
      href = "https://www.hurriyetdailynews.com"+href
      technology_links
      with open("daily_hurriyet_science_Link.txt","a",encoding="utf-8") as file :
        file.write(href+"\n")
```

```python
import concurrent
t1=time.time()
with concurrent.futures.ProcessPoolExecutor() as execut:
  b_res=[execut.submit(science_get_link_creator,i.strip()) for i in science_page_link]
print(time.time()-t1)

t2=time.time()
with concurrent.futures.ProcessPoolExecutor() as execut:
  b_res=[execut.submit(technology_get_link_creator,i.strip()) for i in technology_page_link]
print(time.time()-t2)
```

```python
technology_links = [] 
science_links = []
```

```python
with open("daily_hurriyet_science_Link.txt",'r',newline='') as f:
  for i in f.readlines():    
  # reader=csv.reader(f)
    i = i.strip("\n")
    science_links.append(i)

with open("daily_hurriyet_technology_Link.txt",'r',newline='') as f:
  for i in f.readlines():    
  # reader=csv.reader(f)
    i = i.strip("\n")

    technology_links.append(i)
```

```python
print(technology_links)
print(science_links)
```

```python
def creator_science_content(url):
  html = requests.get(url).content
  soup = BeautifulSoup(html,"html.parser")
  title = soup.find("div",{"class": "content"}).find("h1").getText() # TİTLE Tamam
  date = soup.find("div",{"class" : "content-info"}).find_all("ul")
  date = date[1].getText()

  date = date[:-9]
  date= date.strip()


    
  content = soup.find("div",{"class" : "content"}).find_all("p")
  content_string =""
  categori ="Science"
  
  for i in content:
    i=i.getText().replace(" "," ")
    z=" ".join(i.split())        
    content_string += z
          

  sentence = "{} ; {} ; {} ; {} ; {}".format(url,date,categori,title,content_string)
  with open("hurriyetdaily_Science_Cont.txt","a",encoding="utf-8") as file:
    file.write(sentence +"\n")
```

```python
def creator_technology_content(url):
  html = requests.get(url).content
  soup = BeautifulSoup(html,"html.parser")
  title = soup.find("div",{"class": "content"}).find("h1").getText() # TİTLE Tamam

  date = soup.find("div",{"class" : "content-info"}).find_all("ul")
  date = date[1].getText()
  date = date[:-9]
  date= date.strip()

  content = soup.find("div",{"class" : "content"}).find_all("p")
  content_string =""
  categori ="Technology"
  #if content=="":
  #    content2=soup.find("div",{"class" : "content"}).find_all("text")
   #for i in content2:
    # content=content + i
  
  for i in content:
    i=i.getText().replace(" "," ")
    z=" ".join(i.split())        
    content_string += z
          

  sentence = "{} ; {} ; {} ; {} ; {}".format(url,date,categori,title,content_string)
  with open("hurriyetdaily_technology_Cont.txt","a",encoding="utf-8") as file:
    file.write(sentence +"\n")
```

```python
import concurrent
t1=time.time()
with concurrent.futures.ProcessPoolExecutor() as execut:
  b_res=[execut.submit(creator_science_content,i.strip()) for i in science_links]
print(time.time()-t1)

t2=time.time()
with concurrent.futures.ProcessPoolExecutor() as execut:
  b_res=[execut.submit(creator_technology_content,i.strip()) for i in technology_links]
print(time.time()-t2)
```
